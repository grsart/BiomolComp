## FASTQ to Artifact

# Filtro de sequ√™ncias atrav√©s de par√¢metros de qualidade

Antes de efetivamente utilizar os dados, precisamos remover os ru√≠dos. Para isso, utilizaremos o *plugin* chamado ```DADA2```. Os seguinte passos ser√£o executados:

1. Filtrar e truncar as leituras
2. Encontrar o conjunto mais prov√°vel de sequ√™ncias √∫nicas nas amostras (ASVs)
3. remover chimeras
4. contar a abund√¢ncia de cada ASV

Este comando demorar√° cerca de 2 minutos.

````python
!qiime dada2 denoise-single \
    --i-demultiplexed-seqs sequences.qza \
    --p-trunc-len 150 \
    --p-n-threads 2 \
    --output-dir dada --verbose
````

# Diversidade

## Parte 1: Come√ßando pelo b√°sico

Dentro do pacote `diversity` existe uma a√ß√£o denominada `core-metrics-phylogenetic`. Esta √© responsavel por calcular diversas m√©tricas de diversidade, filogen√©ticas ou n√£o.

Neste ponto, √© importante fazer o uso dos metadados. Assim, podemos rastrear quais amostras v√™m de quais grupos.

Para lidar com a varia√ß√£o na profundidade de amostra, utilizaremos um cutoff que ir√° rarear as amostras de forma **aleat√≥ria**. Desta forma, os dados podem ficar ligeiramente diferentes entre voc√™s.  

````python
!qiime diversity core-metrics-phylogenetic \
    --i-table dada/table.qza \
    --i-phylogeny tree/rooted_tree.qza \
    --p-sampling-depth 5000 \
    --m-metadata-file data/metadata.tsv \
    --output-dir diversity
````

Todas as m√©tricas de diversidade estar√£o salvas na pasta `diversity`. Caso queira maior informa√ß√µes sobre como essas m√©tricas s√£o calculadas e como podem ser intepretadas, verifique o [forum post](https://forum.qiime2.org/t/alpha-and-beta-diversity-explanations-and-commands/2282) do QIIME.


Neste momento, √© importante lembrar da estat√≠stica. Estamos interessados em verificar se a diversidade difere ***significantemente*** entre as amostras de pacientes saud√°veis e doentes, ou alguma outra vari√°vel.

### Diversidade alfa: Riqueza e balanceamento _dentro_ da amostra

Nos comandos anteriores, calculamos tanto a medida de alfa e beta diversidade. Inicialmente, vamos gerar a visualiza√ß√£o da alfa diversidade nas amostras usando o vetor de Shannon.

Em geral, indiv√≠duos saud√°veis e mais velhos possuem uma diversidade balanceada de microbiomas. Isso n√£o necessariamente √© um indicador de sa√∫de ou doen√ßa.

````python
!qiime diversity alpha-group-significance \
    --i-alpha-diversity diversity/shannon_vector.qza \
    --m-metadata-file data/metadata.tsv \
    --o-visualization diversity/alpha_groups.qzv
````

# Taxonomic Classification

### Constru√ß√£o de uma tabela de contagem taxon√¥mica




A pergunta agora √© quais microorganismos est√£o em cada amostra. Utilizaremos classificadores Bayesianos pr√©-treinados. Outros classificadores podem ser encontrados no link: https://docs.qiime2.org/2024.5/data-resources/

Especificamente, utilizaremos um classificador Bayesiano Naive espec√≠fico para a regi√£o V4 do 16S (515f-806r). A raz√£o para essa escolha √© que foi essa a regi√£o sequenciada.

````python
# run a classifier on our 10-sample mini dataset
!qiime feature-classifier classify-sklearn \
    --i-reads dada/representative_sequences.qza \
    --i-classifier ncbi-refseq-genus-515f-806r.qza \
    --p-n-jobs 2 \
    --o-classification taxa.qza
````

Uma vez classificadas as leituras, podemos visualizar a divis√£o taxon√¥mica das amostras

````python
!qiime taxa barplot \
    --i-table dada/table.qza \
    --i-taxonomy taxa.qza \
    --m-metadata-file data/metadata.tsv \
    --o-visualization taxa_barplot.qzv
````

A tabela ```table.qza```, cont√©m as nossas leituras, enquanto o arquivo ```taxa.qza```, cont√©m a classifica√ß√£o taxon√¥mica das leituras. Vamos usar ambas as informa√ß√µes para colapsar os dados a n√≠vel de Genus.

````python
!qiime taxa collapse \
    --i-table dada/table.qza \
    --i-taxonomy taxa.qza \
    --p-level 6 \
    --o-collapsed-table genus.qza
````

Vamos dar uma olhada nos resultados

````python
abundances = pd.read_table("genus.tsv", skiprows=1, index_col=0)
abundances
````

Para deixar mais facil de interpretar, vamos criar uma representa√ß√£o em forma de mapa de calor, utilizando a biblioteca seaborn do python.

````python
import numpy as np
import seaborn as sns
````

````python
abund_to_plot = abundances.copy()
abund_to_plot.index = abund_to_plot.index.str.split(";").str[5]          # Use only the genus name
abund_to_plot = abund_to_plot[~abund_to_plot.index.isin(["g__", "__"])]  # remove unclassified genera
abund_to_plot = abund_to_plot.sample(50, axis=0)                         # use 50 random genera (rows)

# Let's do a centered log-ratio transform: log x_i - log mean(x)
transformed = abund_to_plot.apply(
    lambda xs: np.log(xs + 0.5) - np.log(xs.mean() + 0.5),
    axis=0)

sns.clustermap(transformed.T, cmap="magma", xticklabels=True, figsize=(16, 6))
````

Cada linha √© uma amostra e cada coluna √© um Genus bacteriano. Os valores da tabela s√£o o n√∫mero de vezes que aquele Genus √© detectado em certa amostra.
